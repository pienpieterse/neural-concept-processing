Read results

```python
import pickle

with open("../results/results_blind.pkl", "rb") as f:
    blindresults = pickle.load(f)

with open("../results/best_voxels_ctrlAV.pkl", "rb") as f:
    AVresults = pickle.load(f)

with open("../results/results_ctrlA.pkl", "rb") as f:
    Aresults = pickle.load(f)
```

```python
blindresults
```

```python
import pandas as pd

groups = {
   'blind': blindresults,
   'AV': AVresults,
   'A': Aresults
}

def flatten_results_to_df(groups_dict):
    rows = []

    for group_name, participants in groups_dict.items():

        for participant_id, rois in participants.items():

            for roi_id, model_dict in rois.items():

                for model_name, arr in model_dict.items():

                    # arr is a numpy array  iterate through each value
                    for idx, value in enumerate(arr):
                        rows.append({
                            "group": group_name,
                            "participant": participant_id,
                            "roi": roi_id,
                            "model": model_name,
                            "index": idx,
                            "value": value
                        })

    return pd.DataFrame(rows)

df = flatten_results_to_df(groups)

print(df.head())
```

Add qwentext results

```python
with open("../results/qwentext/results_blind.pkl", "rb") as f:
    blindresultsq = pickle.load(f)

with open("../results/qwentext/results_ctrlAV.pkl", "rb") as f:
    AVresultsq = pickle.load(f)

with open("../results/qwentext/results_ctrlA.pkl", "rb") as f:
    Aresultsq = pickle.load(f)
```

```python
newgroups = {
   'blind': blindresultsq,
   'AV': AVresultsq,
   'A': Aresultsq
}

df_new = flatten_results_to_df(newgroups)

models_to_remove = ['binder_concreteness', 'binder_abstractness']

df_new_filtered = df_new[~df_new['model'].isin(models_to_remove)]

df_all = pd.concat([df, df_new_filtered], ignore_index=True)
```

Add semantic results

```python
with open("../results/semantic/results_semantic.pkl", "rb") as f:
    semanticresults = pickle.load(f)

print(semanticresults)
```

```python
import numpy as np

# The original correlation dict:
# corr_dict = { (binder_model, model_name): array([...]), ... }

corr_conc = {
    model: arr
    for (binder, model), arr in semanticresults.items()
    if binder == 'binder_concreteness'
}

corr_abs = {
    model: arr
    for (binder, model), arr in semanticresults.items()
    if binder == 'binder_abstractness'
}
```

```python
df_all["corr_concreteness"] = np.nan
df_all["corr_abstractness"] = np.nan
```

```python
for model_name, arr in corr_conc.items():
    mask = df_all["model"] == model_name
    df_all.loc[mask, "corr_concreteness"] = arr[df_all.loc[mask, "index"].values]
    
for model_name, arr in corr_abs.items():
    mask = df_all["model"] == model_name
    df_all.loc[mask, "corr_abstractness"] = arr[df_all.loc[mask, "index"].values]
```

```python
df_all
```

Change structure

```python
df_long = df_all.melt(
    id_vars=['group', 'participant', 'roi', 'model', 'index', 'value'],
    value_vars=['corr_concreteness', 'corr_abstractness'],
    var_name='semantic_model',
    value_name='semantic_value'
)

# Optionally clean semantic_model names
df_long['semantic_model'] = df_long['semantic_model'].str.replace('corr_', '')
```

```python
# Create the column
df_long["roi_type"] = df_long["roi"].apply(
    lambda x: "language" if isinstance(x, (int, float)) else "visual"
)

# Move roi_type right after roi
roi_index = df_long.columns.get_loc("roi")
df_long.insert(roi_index + 1, "roi_type", df_long.pop("roi_type"))

```
